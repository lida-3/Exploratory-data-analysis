# Visualization of collected data

For some years there are the same types of data collected through the archive.

This information is shown on the same graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import calendar
import numpy as np
import plotly.express as px
from google.colab import files

```

```python
data1 = pd.read_excel('final2015.xlsx')
data1 = data1.drop(columns=['Unnamed: 0'])

data2 = pd.read_excel('final2016.xlsx')
data2 = data2.drop(columns=['Unnamed: 0'])

data3 = pd.read_excel('final2017.xlsx')
data3 = data3.drop(columns=['Unnamed: 0'])
data4 = pd.read_excel('2018-pp sezona.xlsx')
data4 = data4.drop(columns=['Unnamed: 0'])
data5 = pd.read_excel('2019.xlsx')
data5 = data5.drop(columns=['Unnamed: 0'])
data6 = pd.read_excel('2020.xlsx')
data6 = data6.drop(columns=['Unnamed: 0'])
data7 = pd.read_excel('2021drFormat.xlsx')
data7 = data7.drop(columns=['Unnamed: 0'])
data8 = pd.read_excel('final2022.xlsx')
data8 = data8.drop(columns=['Unnamed: 0'])

```

```python
#change date format
data1['date'] = pd.to_datetime(data1['date'], format="%d.%m.%Y.")
data2['date'] = pd.to_datetime(data2['date'], format="%d.%m.%Y.")
data3['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
data4['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
data5['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
data6['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
data7['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
data8['date'] = pd.to_datetime(data3['date'], format="%d.%m.%Y.")
```

```python
#sort everything from 2018 to 2021
new_data1 = data1.loc[::-1]
new_data2 = data2.loc[::-1]
new_data3 = data3.loc[::-1]
new_data4 = data4.loc[::-1]
new_data5 = data5.loc[::-1]
new_data6 = data6.loc[::-1]
new_data7 = data7.loc[::-1]
new_data8 = data8.loc[::-1]
```

```python
new_data1['date'] = pd.to_datetime(new_data1['date'], format="%Y-%m-%d")
new_data1['month_day'] = new_data1['date'].dt.strftime('%m-%d')
new_data2['date'] = pd.to_datetime(new_data2['date'], format="%Y-%m-%d")
new_data2['month_day'] = new_data2['date'].dt.strftime('%m-%d')
new_data3['date'] = pd.to_datetime(new_data3['date'], format="%Y-%m-%d")
new_data3['month_day'] = new_data3['date'].dt.strftime('%m-%d')
new_data4['date'] = pd.to_datetime(new_data4['date'], format="%Y-%m-%d")
new_data4['month_day'] = new_data4['date'].dt.strftime('%m-%d')
new_data5['date'] = pd.to_datetime(new_data5['date'], format="%Y-%m-%d")
new_data5['month_day'] = new_data5['date'].dt.strftime('%m-%d')
new_data6['date'] = pd.to_datetime(new_data6['date'], format="%Y-%m-%d")
new_data6['month_day'] = new_data6['date'].dt.strftime('%m-%d')
new_data7['date'] = pd.to_datetime(new_data7['date'], format="%Y-%m-%d")
new_data7['month_day'] = new_data7['date'].dt.strftime('%m-%d')
new_data8['date'] = pd.to_datetime(new_data8['date'], format="%Y-%m-%d")
new_data8['month_day'] = new_data8['date'].dt.strftime('%m-%d')
```

##Tons of water used for firefighting in years 2015-2022

```python
new_data1['tons_of_water'] = new_data1['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data2['tons_of_water'] = new_data2['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data3['tons_of_water'] = new_data3['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data4['tons_of_water'] = new_data4['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data5['tons_of_water'] = new_data5['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data6['tons_of_water'] = new_data6['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data7['tons_of_water'] = new_data7['tons_of_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
new_data8['total_tons_water'] = new_data8['total_tons_water'].replace(0, pd.NA).fillna(method='ffill').fillna(0)
```

```python
new_data1['month_day'] = pd.to_datetime(new_data1['month_day'], format='%m-%d')
new_data2['month_day'] = pd.to_datetime(new_data2['month_day'], format='%m-%d')
new_data3['month_day'] = pd.to_datetime(new_data3['month_day'], format='%m-%d')
new_data4['month_day'] = pd.to_datetime(new_data4['month_day'], format='%m-%d')
new_data5['month_day'] = pd.to_datetime(new_data5['month_day'], format='%m-%d')
new_data6['month_day'] = pd.to_datetime(new_data6['month_day'], format='%m-%d')
new_data7['month_day'] = pd.to_datetime(new_data7['month_day'], format='%m-%d')
new_data8['month_day'] = pd.to_datetime(new_data8['month_day'], format='%m-%d')
```

```python
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(new_data1['month_day'], new_data1['tons_of_water'], label="Year 2015", color = "brown")
ax.plot(new_data2['month_day'], new_data2['tons_of_water'], label="Year 2016", color = "pink")
ax.plot(new_data3['month_day'], new_data3['tons_of_water'], label="Year 2017", color = "orange")
ax.plot(new_data4['month_day'], new_data4['tons_of_water'], label="Year 2018", color = "blue")
ax.plot(new_data5['month_day'], new_data5['tons_of_water'], label="Year 2019", color = "red")
ax.plot(new_data6['month_day'], new_data6['tons_of_water'], label="Year 2020", color = "yellow")
ax.plot(new_data7['month_day'], new_data7['tons_of_water'], label="Year 2021", color = "green")
ax.plot(new_data8['month_day'], new_data8['total_tons_water'], label="Year 2022", color = "purple")

ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))

# Set the desired step size or interval for displaying dates
ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.yticks(np.arange(0, 80000, step=5000), fontsize = 14)
plt.xlabel("Dates", fontsize = 16)
plt.ylabel("Tons", fontsize = 16)
plt.title("Tons of water used in years 2015-2022", fontsize = 20)
plt.legend(fontsize = 16)
plt.xticks(fontsize = 14, rotation = 45)
plt.show()
```
